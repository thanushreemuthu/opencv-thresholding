### Question 1

Write a Python program using OpenCV to:

1. Read an image (road_lanes.png) and convert it to **grayscale**.
2. Apply **Global Thresholding** using `cv2.threshold()` with a fixed threshold value.
3. Display the **original grayscale image** and the **global thresholded image** using Matplotlib in a **1 × 2 layout**, each with an appropriate **title**.


```python
# Import
import cv2
import matplotlib.pyplot as plt
```

```python
# Read the image using OpenCV
img = cv2.imread('road.jpeg', 0)
```

```python
# Apply Global Thresholding
ret, thresh = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
```

```python
# Display the image using Matplotlib
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title('Original Grayscale Image')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(thresh, cmap='gray')
plt.title('Global Thresholded Image')
plt.axis('off')

plt.show()
```

### Question 2

Write a Python program using OpenCV to:

1. Read an image (Piano_Sheet_Music.png) and convert it to **grayscale**.
2. Apply **Adaptive Thresholding** (Mean or Gaussian) using `cv2.adaptiveThreshold()`.
3. Display the **original grayscale image** and the **adaptive thresholded image** using Matplotlib in a **1 × 2 layout**, each with an appropriate **title**.


```python
# Import
import cv2
import matplotlib.pyplot as plt
```

```python
# Read the image using OpenCV
img = cv2.imread('road.jpeg', 0)
```

```python
# Apply Adaptive Thresholding (Mean)
adaptive_thresh = cv2.adaptiveThreshold(
    img,
    255,
    cv2.ADAPTIVE_THRESH_MEAN_C,
    cv2.THRESH_BINARY,
    11,
    2
)
```

```python
# Display the image using Matplotlib
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title('Original Grayscale Image')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(adaptive_thresh, cmap='gray')
plt.title('Adaptive Thresholded Image')
plt.axis('off')

plt.show()
```

### Question 3

Write a Python program using OpenCV to:

1. Read an image (road_lanes.png) and convert it to **grayscale**.
2. Apply **Otsu’s Thresholding** using `cv2.threshold()` with the `cv2.THRESH_OTSU` flag.
3. Display the **original grayscale image** and the **Otsu-thresholded image** using Matplotlib in a **1 × 2 layout**, each with an appropriate **title**.


```python
# Import
import cv2
import matplotlib.pyplot as plt
```

```python
# Read the image using OpenCV
img = cv2.imread('road.jpeg', 0)
```

```python
# Apply Otsu's Thresholding
ret, otsu_thresh = cv2.threshold(
    img,
    0,
    255,
    cv2.THRESH_BINARY + cv2.THRESH_OTSU
)
```

```python
# Display the image using Matplotlib
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title('Original Grayscale Image')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(otsu_thresh, cmap='gray')
plt.title("Otsu's Thresholded Image")
plt.axis('off')

plt.show()
```
